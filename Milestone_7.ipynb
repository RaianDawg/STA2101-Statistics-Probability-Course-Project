```python
# =========================================================
# Milestone 7: Simple Linear Regression
# Dataset: Global Terrorism Database
# Perspective: Statistician / Security Data Analyst
# =========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# CELL 1: Load Dataset
# ---------------------------------------------------------
df = pd.read_csv("globalterrorismdb_0718dist.csv", encoding='latin1')
df.info()

# ---------------------------------------------------------
# CELL 2: Select Relevant Numerical Columns
# ---------------------------------------------------------
numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
numerical_columns

# ---------------------------------------------------------
# CELL 3: Define Regression Scope (CONTROLLED EXPANSION)
# ---------------------------------------------------------
# Dependent variable (outcome)
Y_vars = ["nkill"]

# Independent variables (attack characteristics)
X_vars = [
    "nwound",
    "nperps",
    "success",
    "suicide",
    "extended",
    "imonth"
]

# Keep only valid columns
X_vars = [x for x in X_vars if x in df.columns]
Y_vars = [y for y in Y_vars if y in df.columns]

# Generate regression pairs
pairs = []
for x in X_vars:
    for y in Y_vars:
        pairs.append((x, y))

print(f"Total regression models: {len(pairs)}")
pairs

# ---------------------------------------------------------
# CELL 4: Simple Linear Regression Analysis
# ---------------------------------------------------------
for X_name, Y_name in pairs:

    print("\n" + "="*80)
    print(f"SIMPLE LINEAR REGRESSION: {Y_name} vs {X_name}")
    print("="*80)

    # --- TASK 1: Data Selection and Initial Visualization ---
    X = df[X_name]
    Y = df[Y_name]

    # Drop missing values
    temp = pd.concat([X, Y], axis=1).dropna()
    X = temp[X_name]
    Y = temp[Y_name]

    print("\n--- Task 1: Data Selection and Initial Visualization ---")
    print(f"Independent Variable (X): {X_name}")
    print(f"Dependent Variable (Y): {Y_name}")

    X_mean = X.mean()
    Y_mean = Y.mean()

    print(f"Mean of X: {X_mean:.4f}")
    print(f"Mean of Y: {Y_mean:.4f}")

    # Scatter plot
    plt.figure(figsize=(8, 5))
    plt.scatter(X, Y, alpha=0.7)
    plt.xlabel(X_name)
    plt.ylabel(Y_name)
    plt.title(f"{Y_name} vs {X_name}")
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()

    # -----------------------------------------------------
    # --- TASK 2: Manual Calculation of Regression Parameters
    # -----------------------------------------------------
    print("\n--- Task 2: Manual Calculation of Regression Parameters ---")

    X_dev = X - X_mean
    Y_dev = Y - Y_mean

    numerator = (X_dev * Y_dev).sum()
    denominator = (X_dev ** 2).sum()

    beta_1 = numerator / denominator
    beta_0 = Y_mean - beta_1 * X_mean

    print(f"Slope (β1): {beta_1:.4f}")
    print(f"Intercept (β0): {beta_0:.4f}")
    print(f"Regression Equation: Ŷ = {beta_0:.4f} + {beta_1:.4f}X")

    # -----------------------------------------------------
    # --- TASK 3: Visualization of Regression Fit
    # -----------------------------------------------------
    print("\n--- Task 3: Visualization of the Fit ---")

    Y_pred = beta_0 + beta_1 * X

    # Sort for clean line
    sort_idx = np.argsort(X)
    X_sorted = X.iloc[sort_idx]
    Y_pred_sorted = Y_pred.iloc[sort_idx]

    plt.figure(figsize=(8, 5))
    plt.scatter(X, Y, alpha=0.7, label="Observed Data")
    plt.plot(X_sorted, Y_pred_sorted, linewidth=2, label="Regression Line")
    plt.xlabel(X_name)
    plt.ylabel(Y_name)
    plt.title(f"Regression Fit: {Y_name} vs {X_name}")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()

    # -----------------------------------------------------
    # --- TASK 4: Strength of Relationship
    # -----------------------------------------------------
    print("\n--- Task 4: Strength of Relationship ---")

    sum_sq_Y = (Y_dev ** 2).sum()
    r = numerator / np.sqrt(denominator * sum_sq_Y)
    R_squared = r ** 2

    print(f"Pearson Correlation (r): {r:.4f}")
    print(f"Coefficient of Determination (R²): {R_squared:.4f}")

    # -----------------------------------------------------
    # --- OPTIONAL: Scikit-learn Validation
    # -----------------------------------------------------
    try:
        from sklearn.linear_model import LinearRegression

        X_skl = X.values.reshape(-1, 1)
        model = LinearRegression()
        model.fit(X_skl, Y)

        print("\n--- OPTIONAL VALIDATION (scikit-learn) ---")
        print(f"Manual β1: {beta_1:.4f} | SKL β1: {model.coef_[0]:.4f}")
        print(f"Manual β0: {beta_0:.4f} | SKL β0: {model.intercept_:.4f}")
        print(f"Manual R²: {R_squared:.4f} | SKL R²: {model.score(X_skl, Y):.4f}")

    except:
        pass
```

## Pair Selection Justification

The regression analysis in this notebook focused on understanding the individual linear relationship between each selected independent variable and the dependent variable, `nkill` (number of confirmed fatalities).

**Dependent Variable:**
*   `nkill`: Selected as the sole dependent variable, representing the outcome we aim to predict or explain.

**Independent Variables:**
*   `nwound`
*   `nperps`
*   `success`
*   `suicide`
*   `extended`
*   `imonth`

The choice was to perform a series of *simple linear regressions*, meaning each model considers only one independent variable at a time against the dependent variable. Therefore, the pairs for analysis were systematically generated by combining each independent variable from the `X_vars` list with the single dependent variable `nkill` from the `Y_vars` list. This approach allows for a clear, isolated examination of the linear correlation and predictive power of each individual characteristic on the number of fatalities, without the confounding effects of other predictors. This generates pairs such as (`nwound`, `nkill`), (`nperps`, `nkill`), and so on, for each independent variable.

## Analysis and Reflection

This notebook performed a series of simple linear regression analyses to explore the relationship between various attack characteristics and the number of fatalities (`nkill`) in the Global Terrorism Database. The goal was to identify which single independent variables, among those selected, showed the strongest linear correlation with `nkill`.

**Interpretation & Summary of Results:**

1.  **`nkill` vs. `nwound` (Number Wounded):**
    *   **Pearson Correlation (r):** 0.5344
    *   **Coefficient of Determination (R²):** 0.2856
    *   **Equation:** `Ŷ = 1.6705 + 0.1456X`
    *   **Interpretation:** This model showed the strongest linear relationship among all tested pairs. Approximately 28.56% of the variance in `nkill` can be explained by `nwound`. This suggests a moderate positive correlation: as the number of wounded increases, the number of kills tends to increase. This is an intuitive finding, as more severe attacks generally result in both more wounded and more fatalities.

2.  **`nkill` vs. `nperps` (Number of Perpetrators):**
    *   **Pearson Correlation (r):** 0.0250
    *   **Coefficient of Determination (R²):** 0.0006
    *   **Equation:** `Ŷ = 2.5545 + 0.0015X`
    *   **Interpretation:** An extremely weak positive correlation. The number of perpetrators explains less than 0.1% of the variance in `nkill`. This indicates that simply knowing the number of perpetrators has almost no linear predictive power for the number of kills.

3.  **`nkill` vs. `success` (Attack Success):**
    *   **Pearson Correlation (r):** 0.0531
    *   **Coefficient of Determination (R²):** 0.0028
    *   **Equation:** `Ŷ = 0.6995 + 1.9245X`
    *   **Interpretation:** A very weak positive correlation. Only 0.28% of the variance in `nkill` is explained by the success of an attack. While logically one might expect successful attacks to lead to more kills, this simple linear model suggests that 'success' as a binary variable (`0` or `1`) does not linearly predict the magnitude of fatalities well on its own.

4.  **`nkill` vs. `suicide` (Suicide Attack):**
    *   **Pearson Correlation (r):** 0.1361
    *   **Coefficient of Determination (R²):** 0.0185
    *   **Equation:** `Ŷ = 2.0907 + 8.2102X`
    *   **Interpretation:** A weak positive correlation. Suicide attacks explain about 1.85% of the variance in `nkill`. This indicates that suicide attacks, while generally associated with higher casualties, do not have a strong *linear* predictive relationship with the number of kills when considered in isolation.

5.  **`nkill` vs. `extended` (Extended Duration):**
    *   **Pearson Correlation (r):** 0.0277
    *   **Coefficient of Determination (R²):** 0.0008
    *   **Equation:** `Ŷ = 2.3417 + 1.7269X`
    *   **Interpretation:** A very weak positive correlation, explaining less than 0.1% of the variance in `nkill`. The duration of an attack, as captured by the `extended` binary variable, has negligible linear predictive power for fatalities.

6.  **`nkill` vs. `imonth` (Month of Attack):**
    *   **Pearson Correlation (r):** 0.0035
    *   **Coefficient of Determination (R²):** 0.0000
    *   **Equation:** `Ŷ = 2.3268 + 0.0118X`
    *   **Interpretation:** An extremely weak correlation, explaining almost 0% of the variance in `nkill`. The month in which an attack occurs has no discernible linear relationship with the number of kills.

**Comparison:**
Comparing the R-squared values, `nwound` stands out significantly as the best single linear predictor of `nkill` among the selected variables. All other variables (`nperps`, `success`, `suicide`, `extended`, `imonth`) exhibited very weak to negligible linear relationships, explaining less than 2% of the variance in `nkill` individually.

**Reflection & Conclusion:**

The simple linear regression models highlight that `nwound` has a non-trivial, though still moderate, linear relationship with `nkill`. This suggests that the severity of an attack (as indicated by wounded) is a significant correlate of its lethality. However, the low R-squared values for most other variables indicate that simple linear regression, using single attack characteristics, is largely insufficient to comprehensively explain the variability in `nkill`.

Terrorism fatalities are likely influenced by a complex interplay of many factors, including target type, weapon type, group motivation, country context, and more. Future analyses would benefit from exploring multiple linear regression, non-linear models, or more sophisticated machine learning approaches to capture these intricate relationships. The current analysis serves as a foundational step, identifying `nwound` as a primary driver, but also underscoring the multivariate nature of terrorism fatalities.
